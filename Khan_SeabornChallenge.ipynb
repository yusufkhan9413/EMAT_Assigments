# Challenge for group 1

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Loading data

titanic = pd.read_csv('titanic_clean.csv')
titanic.head(1)
```

```python
# plotting relplot

sns.relplot(
    data = titanic,
    x = "age",
    y = "fare"
    )

plt.show()      # Can also use it for plotting

# seaborn.object is used to manuplate the objects of the plot
```

```python
# alpha shows us the transperancy. 1 is the opaque, 0 is total transperancy

sns.relplot(
    data = titanic,
    x = "age",
    y = "fare",
    alpha = 0.5
)
```

```python
# in "hue" we provide the name of the column on which color has to be decided

sns.relplot(
    data = titanic,
    x = "age",
    y = "fare",
    alpha = 0.5,
    hue = "sex"
)
```

```python
# in "col" is used to add a new column in the plot

sns.relplot(
    data = titanic,
    x = "age",
    y = "fare",
    alpha = 0.5,
    hue = "sex",
    col = "survived"
)
```

```python
# palette is used to give color to different 

sns.relplot(
    data = titanic,
    x = "age",
    y = "fare",
    alpha = 0.5,
    hue = "sex",
    col = "survived",
    palette = {"male": "blue", "female": "magenta"}
)
```

```python
# way to change the name of the column to show in the graph

titanic['Sex'] = titanic['sex'].map({"male": "Male", "female": "Female"})
```

```python
sns.relplot(
    data = titanic,
    x = "age",
    y = "fare",
    alpha = 0.5,
    hue = "Sex",
    col = "survived",
    palette = {"Male": "blue", "Female": "magenta"}
)
```

```python
sns.relplot(
    data = titanic,
    x = "age",
    y = "fare",
    alpha = 0.5,
    hue = "Sex",
    col = "survived",
    palette = {"Male": "blue", "Female": "magenta"},
    size = "pclass"
)
```

```python
titanic['survived_ind'] = titanic['survived'].map({0: "did not survive", 1: "survived"})
```

```python
sns.relplot(
    data = titanic,
    x = "age",
    y = "fare",
    alpha = 0.5,
    hue = "survived_ind",
    col = "survived",
    palette = {"did not survive": "#066F8E", "survived": "#76161D"}
)
```

# Challenge for Group 2

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# reading the csv file

smh = pd.read_csv("Student Mental health.csv")
smh.head(1)
```

```python
# Groupby using the gender and mental health

smh.groupby("Choose your gender")[["Do you have Depression?", "Do you have Anxiety?", "Do you have Panic attack?"]].count()
```

```python
# Selecting the last 4 and then excluding the last column

columns_to_plot = smh.columns[-4:-1]

category_order = ['Yes', 'No']
```

```python
for column in columns_to_plot:
    counts = smh[column].value_counts()     # value_count will give us values in the alphabetical order.     Output might have been No and Yes
    counts = counts.reindex(category_order, fill_value=0)  # reindex wll take the values out of alphabetical order to what we want it to be.    Now output is Yes and No
    print(counts)
    print()
```

```python
# generating the plot for Depression

sns.countplot(data= smh, 
              x = 'Do you have Depression?', 
              stat = 'percent', 
              hue = 'Did you seek any specialist for a treatment?'
             )
```

```python
# generating the plot for Anxiety

sns.countplot(data = smh, 
              x = 'Do you have Anxiety?', 
              stat = 'percent', 
              hue = 'Did you seek any specialist for a treatment?'
             )
```

```python
# generating the plot for Panic attack

sns.countplot(data = smh, 
              x = 'Do you have Panic attack?', 
              stat = 'percent', 
              hue = 'Did you seek any specialist for a treatment?'
             )
```

```python
# Palette PuBuGn is a preset for color in seaborn

sns.countplot(data = smh, 
              x = 'Do you have Depression?', 
              stat = 'percent', 
              hue = 'Choose your gender', 
              palette = 'PuBuGn'
             )
```

```python
sns.countplot(data = smh, 
              x = 'Do you have Anxiety?', 
              stat = 'percent', 
              hue = 'Choose your gender', 
              palette = 'PuBuGn'
             )
```

```python
sns.countplot(data = smh, 
              x = 'Do you have Panic attack?', 
              stat = 'percent', 
              hue = 'Choose your gender', 
              palette = 'PuBuGn'
             )
```

```python
# Another way of doing these same plots

for column in columns_to_plot:
    sns.countplot(data= smh, 
                  x = column, 
                  stat = 'percent', 
                  hue = 'Did you seek any specialist for a treatment?'
                  )
    plt.show()
```
