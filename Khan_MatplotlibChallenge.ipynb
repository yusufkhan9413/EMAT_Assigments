# HeatMap

They help identify trends, patterns, and anomalies within data, such as correlations between different metrics or the frequency of events.

**Interpretation of Correlation Heatmaps:**

**Positive Correlation**: This indicates that as one variable increases and the other also increases. Shown in warmer colors (reds or oranges).

**Negative Correlation**: This indicates that as one variable increases, the other decreases. Shown in cooler colors (blue).

**Weak or No Correlation**: This indicates no strong linear relationship between the variables. These are the correlations close to zero and shown in neutral colors (like white or light yellow)

Examples of usage of Heatmap: Insurance, Finance, Healthcare, Supply Chain, Prediction.hain

```python
#creating a scatter plot 
fig2,axes2 = plt.subplots()
axes2.scatter(x=df['ConMediaUse'], y=df['ImmiPhobia'], color='lightblue', marker='o')
#adding labels and a title 
axes2.set_xlabel('ConMediaUse')
axes2.set_ylabel('ImmiPhobia')
axes2.set_title('Scatter Plot of ConMediaUse vs. ImmiPhobia')
plt.show()
```

# Ways to improve the scatter pots:

**Partial Transperancy**: One way to overlapping point problem is to use partial transparency. If we make individual points partially transparent, then overplotted points appear as darker points and thus the shade of the points reflects the density of points in that location of the graph. 

**Jittering**: A simple trick that can further helps is to apply a small amount of jitter to the points, i.e., to displace each point randomly by a small amount in either the x or the y direction or both. With jitter, it is immediately apparent that the darker areas arise from points that are plotted on top of each other

**Interactive Scatter Plots**: For larger or more complex datasets, turning the scatterplot into an interactive visualization can greatly enhance usability. Viewers can hover over individual points to see specific values, filter data dynamically, or zoom in on areas of interest.

**Aesthetics**: Aesthetics describe every aspect of a given graphical element such as a shape, a size, and a color. We can add different shapes and colors for different variables.


**Downside of Jittering**: One downside of jittering is that it does change the data and therefore has to be performed with care. If we jitter too much, we end up placing points in locations that are not representative of the underlying dataset. The result is a misleading visualization of the data.
